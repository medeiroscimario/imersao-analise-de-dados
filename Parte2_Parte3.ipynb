```python
# Agrupamento
df.groupby("tipo")["vitimas"].sum().to_frame().sort_values(by="vitimas", ascending=False)
```

```python
# Removendo da visualização dos dados
tipos = ["APOIO EMLURB", "FISCALIZAÇÃO", "MONITORAMENTO", "PROTESTO", "RENDIÇÃO"]
```

```python
# O isin pega tudo o que está na lista, quando eu passo o ~, ele acaba invertendo, e pegando tudo menos o que está na lista
tp = df[~df['tipo'].isin(tipos)]
```

```python
tp['tipo'].unique()
```

```python
# Top 10 acidentes por bairro
top_10 = df["bairro"].value_counts().head(10).reset_index().sort_values(by='count', ascending=True)
```

```python
fig = px.bar(top_10, x="bairro", y="count", title="Top 10 acidentes por bairro", text="count",
             color_discrete_sequence=['#D2691E'])
fig.show()
```

```python
fig = px.bar(top_10, x="count", y="bairro", title="Top 10 acidentes por bairro", text="count",
             color_discrete_sequence=['#D2691E'])
fig.show()
```

```python
# Total de acidentes por mês
(df["data"].dt.month).value_counts()
```

```python
#Quando se tem uma coluna com elementos do tipo datetime, eu posso extrair dia, mês, anos
df["data"].dt.month
```

```python
# Adicionar uma nova coluna no DataFrame
df['mes_acidente'] = df["data"].dt.month
```

```python
df.head(1)
```

```python
total_mes = df['mes_acidente'].value_counts().reset_index()
total_mes = total_mes.sort_values(by='mes_acidente', ascending=True)
total_mes["mes_acidente"] = total_mes["mes_acidente"].astype(str)
total_mes
```
